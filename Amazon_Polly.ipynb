```python
text="OpenInApp is an innovative software product that seamlessly bridges the gap between web and mobile applications, allowing users to easily open links from the web directly in their preferred mobile apps, enhancing the overall user experience and streamlining the transition between online and mobile interactions."
text
```

```python
response = polly.synthesize_speech(Text=text, OutputFormat="mp3",
                                    VoiceId="Amy")
if "AudioStream" in response:
        with response["AudioStream"] as stream:
           output_file = "speech.mp3"
           try:
            # Open a file for writing the output as a binary stream
                with open(output_file, "wb") as file:
                   file.write(stream.read())
           except IOError as error:
              # Could not write to file, exit gracefully
              print(error)
else:
    # The response didn't contain audio data, exit gracefully
    print("Could not stream audio")
```
